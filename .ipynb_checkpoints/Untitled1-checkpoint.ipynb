```python
titles = pd.read_excel('DataStory/Bitcoin_data/BITCOIN-BTCEEUR.xls', index_col=None)
titles.columns
```

```python
date = titles['Date']
currency = titles['Volume (Currency)']
print (max(currency))
currency = currency.div(currency.loc[0])

btc_currency = titles['Volume (BTC)']
btc_currency = btc_currency.div(btc_currency.loc[0])
print (max(btc_currency))

fig, ax = plt.subplots()

ax.plot( date, currency , label = 'Currency')
ax.plot( date, btc_currency, label = 'Bitcoin' )
ax.legend(loc=2)
ax.set_xlabel('Date')
ax.set_ylabel('Currency')
ax.set_title('Currency/Bitcoin vs Date')

diff = currency - btc_currency
fig, ax1 = plt.subplots()
ax1.plot(date, diff, label = 'Trade diff')
ax1.set_xlabel('Date')
ax1.set_ylabel('Trade Diff')
ax1.set_title('TradeDiff vs Date')
```
