```python
# List the top 5 regions for startups in terms of investment.
investments = pd.DataFrame.from_csv('DataStory/startups/crunchbase-investments.csv', index_col=None)
investments.company_region.value_counts().head()
```

```python
# Percentage of companies whom got the investors from the same city. 
df = investments[investments.company_city == investments.investor_city]
print (" % of companies having investors from same city is : " + str(len(df)))
investments.groupby(['company_city', 'investor_city']).size().plot(kind = 'bar')
```

```python
# Percenatge of companies who got investors from the same region. 
df = investments[investments.company_region == investments.investor_region]
print (" % of companies having investors from same region is : " + str(len(df)))
investments.groupby(['company_region', 'investor_region']).size().plot(kind = 'bar')
```

```python
# Percentage of companies who have investors from the same state 
df = investments[investments.company_state_code == investments.investor_state_code]
print (" % of companies having investors from same state is : " + str(len(df)))

investments.groupby(['company_state_code', 'investor_state_code']).size().plot(kind = 'bar')
```

```python
# state wise plot of investors , 48 investments done from MA 
investments = investments[investments.investor_state_code.notnull()]
investments.investor_state_code.value_counts().plot( kind = 'bar')
```

```python
# plot of investors by name. A single investment group has made all 48 investments. 
investments = investments[investments.investor_name.notnull()]
investments.investor_name.value_counts().plot(kind = 'bar')
```

```python
# Plot of investments made by .406 securities according to market.
investor_406 = investments[investments.investor_name == '.406 Ventures']

investor_406.groupby(['company_market']).size().plot( kind = 'bar')
```

```python
# Plot of investments made by .406 according to region 
investor_406.groupby(['company_region']).size().plot( kind = 'bar')
```

```python
# Plot of investors and companies being from the same market. 
df = investments[investments.company_market == investments.investor_market]
print (" % of companies having investors from same market is : " + str(len(df)))
investments.groupby(['investor_market','company_market']).size().plot(kind = 'bar')
```
